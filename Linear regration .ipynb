#  Standardize or Feature scalling the dataset 

```python
## standardize or feature scalling the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler
```

```python
scaler ## Scaler is an object, Standard Scaler Objects
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

#  Model Training 

## Linear Regression Model

```python
## Linear Regression 
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(x_train, y_train)
```

### Print the coefficients and the intercept

```python
## print the coefficients
print(regression.coef_)
```

```python
## print intercept
print(regression.intercept_)
```

```python
## prediction for the test data 
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

# Assumption of Linear Regression 

```python
## Assumption of Linear Regression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Relationship between real data and predict data
plt.scatter(y_test,reg_pred)
plt.xlabel("test truth data")
plt.ylabel("test predicted data")
```

```python
## calculating Residual
residuals = y_test - reg_pred
```

```python
residuals
```

```python
## Distribution of residual are approximately normal distribution
sns.displot(residuals, kind ="kde")
```

```python
## Scatter plot with predictions and residual
## Unifrom distributions
plt.scatter(reg_pred, residuals) ## unifrom Distribution model
```

## Performance Metrics

```python
## performance Matrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error (y_test, reg_pred))
print(mean_absolute_error (y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

## R Square and adjusted R Squared

```python
## R squared 
from sklearn.metrics import r2_score
score=r2_score(y_test, reg_pred)
print(score)
```

```python
## adjested r Square
## display r adjusted r-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
